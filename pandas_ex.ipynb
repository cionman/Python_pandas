```python
t1 = pd.DataFrame({
    "id" : [1,2,3,4,5],
    "left_val" : ["a", "b", "c","d","e"]
})
t1
```

```python
t2 = pd.DataFrame({
    "id" : [3,4,5,6,7],
    "right_val" : ["a", "w", "e", "r", "t"]
})
```

```python
# merge는 SQL의 Join
pd.merge(t1, t2)
```

```python
# 특정 열 기준 join, on 키워드 파라미터
pd.merge(t1, t2, on="id")
```

```python
# 양쪽 기준이 다를 때
pd.merge(t1, t2, left_on="left_val", right_on = "right_val")
```

```python
# left outer join
pd.merge(t1, t2, how='left')
```

```python
# right outer join
pd.merge(t1,t2, how='right')
```

```python
# full outer join
pd.merge(t1, t2, how='outer')
```

```python
by_country.get_group('South Korea')
```

```python
by_country.agg({
    'level' : [("합계", np.sum), np.mean]
})
```

```python
def sum_of_pow(s:pd.core.series.Series):
    return sum([i ** 2 for i in s])

by_country.agg({
    "level" : [("sum_of_pow",sum_of_pow)]
}).tail()
```

```python
def get_diff_date(df:pd.core.frame.DataFrame):
    return df.df_date.max() - df.df_date.min()
by_country.apply(get_diff_date)
```
